```python
%matplotlib inline
import matplotlib.image as mpimg
im1=mpimg.imread('53889020.jpg')
im2=mpimg.imread('54217261.jpg')
im3=mpimg.imread('54353990.jpg')
im4=mpimg.imread('54729174.jpg')
im5=mpimg.imread('55457292.jpg')
import matplotlib.pyplot as plt
plt.imshow(im1)
import matplotlib.pyplot as plt
plt.imshow(im2)
```

```python
b_im1=im1[:,:,2]
print(b_im1)
print(b_im1.shape)
```

```python
import numpy as np
b_im2=im2[:,:,2]
b_im3=im3[:,:,2]
b_im4=im4[:,:,2]
b_im5=im5[:,:,2]
b_im=np.stack((b_im1,b_im2,b_im3,b_im4,b_im5))
print(b_im.shape)
print(b_im)
```

```python
print(np.mean(b_im))
b_ave=np.mean(b_im,axis=0)
plt.imshow(b_ave)
```

```python
b_med=np.median(b_im,axis=0)
plt.imshow(b_med)
```

```python
im=np.stack((im1,im2,im3,im4,im5),axis=0)
im_ave=np.mean(im,axis=0)
print(im_ave.shape)
plt.imshow(im_ave.astype(np.uint8)     
plt.savefig("one_5th_of.jpg",bbox_inches='tight',pad_inches=0)
```

```python
im_med=np.median(im,axis=0)
print(im_med.shape)
plt.imshow(im_med.astype(np.uint8))
```
